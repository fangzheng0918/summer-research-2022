```julia
using Pkg
Pkg.add("Colors")
using Colors
```

```julia
x = 9 #number of rows
y = 9 #number of columns
A = rand(0:1,x,y) #a random martix represent counters by positions
```

```julia
function num2col(i)
    if i == 1
        return colorant"black"
    elseif i == 0
        return colorant"white"
    end
end
```

```julia
num2col.(A)
```

```julia
B = zeros(Int64, x+2, y+2) #create an empty martix with one more layer of zeros
```

```julia
#move the original random martix into this bigger one
for i in 1:x, j in 1:y
    B[i+1,j+1] = A[i,j]# every counter go 1 down and 1 rifht forward.
end
B
```

```julia
N = zeros(Int64, x, y) #create an empty number of corresponding neighbors martix
```

```julia
#generate the number of corresponding neighbors
for i in 1:x, j in 1:y
    N[i,j]= B[i,j]+B[i,j+1]+B[i,j+2]+B[i+1,j]+B[i+1,j+2]+B[i+2,j]+B[i+2,j+1]+B[i+2,j+2]# Sum the surrounding neighbors up bu locations
end
N
```

```julia
#run the three laws of life
for i in 1:x, j in 1:y
    #survival law:
    if A[i,j] == 1
        if N[i,j] == 2 || N[i,j] == 3# if there are 2 or 3 neighbors
            A[i,j] = 1# then this counter remain alive
        #death law:
        elseif N[i,j] >= 4 || N[i,j] <= 1# if there are greater than or equal to 4, or less than or euqal to 1 neighbors
            A[i,j] = 0# then this counter die
        end
    else
        #birth law:
        if N[i,j] == 3# if this counter have exacut 3 neighbors
            A[i,j] = 1# then generate a new one in this box
        end
    end
end
A
```

```julia
#make a function of generate next martix based on pervious one
function runlaws(A)
    for i in 1:x, j in 1:y
        B[i+1,j+1] = A[i,j]
    end
    N = zeros(Int64, x, y) #create an empty number of corresponding neighbors martix
    #generate the number of corresponding neighbors
    for i in 1:x, j in 1:y
        N[i,j]= B[i,j]+B[i,j+1]+B[i,j+2]+B[i+1,j]+B[i+1,j+2]+B[i+2,j]+B[i+2,j+1]+B[i+2,j+2]
    end
    #run the three laws of life
    for i in 1:x, j in 1:y
        #survival
        if A[i,j] == 1
            if N[i,j] == 2 || N[i,j] == 3
                A[i,j] = 1
            #death
            elseif N[i,j] >= 4 || N[i,j] <= 1
                A[i,j] = 0
            end
        else
            #birth
            if N[i,j] == 3
                A[i,j] = 1
            end
        end
    end
    return A
end
```

```julia
#run this fucntion
runlaws(A)
```
