# Generate substitution by square tiling

```julia
#We need to generate a tiling first by using 1s and 0s
A = rand(0:1,8,8)
```

```julia
#give all of these suqare a number in order from 1 to the end
count = 1 
for i in 1:8, j in 1:8
    if A[i,j] == 1
        A[i,j] = count
        count +=1
    end
end
A
```

```julia
#Create a dictionary by putting all of the suqares in
adj_dict = Dict()

for i in 1:8, j in 1:8
   tiling_number = A[i,j]
   if tiling_number != 0
        adj_dict[tiling_number] = [
            [0,0,0,0],
            ["U","R","D","L"]
        ]      
    end
end    
adj_dict
```

```julia
#Check all of these tiles whether or not on the boundaries
#Check the potential neighbors of these tiles
for i in 1:8, j in 1:8
    tiling_number = A[i,j]
        if tiling_number != 0
                if i== 1 && j == 1
                    adj_dict[tiling_number][1] = [0,0]
                    adj_dict[tiling_number][2] = ["R", "D"]
                    if A[1,2] != 0
                        adj_dict[tiling_number][1][1] = A[1,2]
                    end
                    if A[2,1] != 0
                            adj_dict[tiling_number][1][2] = A[2,1]
                    end

                elseif i == 1 && j == 8
                    adj_dict[tiling_number][1] = [0,0]
                    adj_dict[tiling_number][2] = ["D", "L"]
                    if A[2,8] != 0
                        adj_dict[tiling_number][1][1] = A[2,8]
                    end
                    if A[1,7] != 0
                        adj_dict[tiling_number][1][2] = A[1,7]
                    end

                elseif i == 8 && j == 1
                    adj_dict[tiling_number][1] = [0,0]
                    adj_dict[tiling_number][2] = ["U", "R"]
                    if A[7,1] != 0
                        adj_dict[tiling_number][1][1] = A[7,1]
                    end
                    if A[8,2] != 0
                        adj_dict[tiling_number][1][2] = A[8,2]
                    end

                elseif i == 8 && j == 8
                    adj_dict[tiling_number][1] = [0,0]
                    adj_dict[tiling_number][2] = ["U", "L"]
                    if A[7,8] != 0
                        adj_dict[tiling_number][1][1] = A[7,8]
                    end
                    if A[8,7] != 0
                        adj_dict[tiling_number][1][2] = A[8,7]
                    end

                elseif i == 1
                    adj_dict[tiling_number][1] = [0,0,0]
                    adj_dict[tiling_number][2] = ["R", "D", "L"]
                    if A[1,j+1] != 0
                        adj_dict[tiling_number][1][1] = A[1,j+1]
                    end
                    if A[2,j] != 0
                        adj_dict[tiling_number][1][2] = A[2,j]
                    end
                    if A[1,j-1] != 0
                        adj_dict[tiling_number][1][3] = A[1,j-1]
                    end

                elseif i== 8
                    adj_dict[tiling_number][1] = [0,0,0]
                    adj_dict[tiling_number][2] = ["U", "R", "L"]
                    if A[7,j] != 0
                        adj_dict[tiling_number][1][1] = A[7,j]
                    end
                    if A[8,j+1] != 0
                        adj_dict[tiling_number][1][2] = A[8,j+1]
                    end
                    if A[8,j-1] != 0
                        adj_dict[tiling_number][1][3] = A[8,j-1]
                    end

                elseif j == 1
                    adj_dict[tiling_number][1] = [0,0,0]
                    adj_dict[tiling_number][2] = ["U", "R", "D"]
                    if A[i-1,1] != 0
                        adj_dict[tiling_number][1][1] = A[i-1,1]
                    end
                    if A[i,2] != 0
                        adj_dict[tiling_number][1][2] = A[i,2]
                    end
                    if A[i+1,1] != 0
                        adj_dict[tiling_number][1][3] = A[i+1,1]
                    end

                elseif j == 8
                    adj_dict[tiling_number][1] = [0,0,0]
                    adj_dict[tiling_number][2] = ["U", "D", "L"]
                    if A[i-1,8] != 0
                        adj_dict[tiling_number][1][1] = A[i-1,8]
                    end
                    if A[i+1,8] != 0
                        adj_dict[tiling_number][1][2] = A[i+1,8]
                    end
                    if A[i,7] != 0
                        adj_dict[tiling_number][1][3] = A[i,7]
                    end
            
                elseif i != 1||8 && j != 1||8
                    if A[i-1,j] != 0
                        adj_dict[tiling_number][1][1] = A[i-1,j]
                    end
                    if A[i,j+1] != 0
                        adj_dict[tiling_number][1][2] = A[i,j+1]
                    end
                    if A[i+1,j] != 0
                        adj_dict[tiling_number][1][3] = A[i+1,j]
                    end
                    if A[i,j-1] != 0
                        adj_dict[tiling_number][1][4] = A[i,j-1]
                    end

                end
        end
end

adj_dict
```

```julia
#remove the Unnecessary 0s and orientations from dict
for i in 1:length(adj_dict)
    tile_arr = adj_dict[i][1]
    orient_arr = adj_dict[i][2]
    
    zero_indices = copy(findall(iszero, tile_arr))
    
    deleteat!(orient_arr,zero_indices)
    deleteat!(tile_arr,zero_indices)

end

adj_dict
```
